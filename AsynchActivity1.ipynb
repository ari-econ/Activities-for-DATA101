```python
import pandas as pd

df = pd.read_excel('EDB_regression_data.xlsx')
print(df.head())
```

```python
X = df[['EDB']]
Y = df['GDP growth (annual %)']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
print(f"Intercept: {model.intercept_}")
print(f"Coefficient: {model.coef_}")
```

```python
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
import matplotlib.pyplot as plt

# Plot the scatter plot
plt.scatter(X, Y, color='#1fb462', alpha=0.6, label="Data points")

# Plot the regression line
plt.plot(X_test, Y_pred, color='#243bad', label="-0.03149")

# Add labels and title
plt.xlabel("Ease of Doing Business (EDB)")
plt.ylabel("GDP Growth (Annual %)")
plt.title("Regression between EDB and GDP Growth (2020)")
plt.legend()
plt.show()
```

```python
# Define X (independent variable) and Y (dependent variable)
X = df[['EDB']]  # Use double brackets for a single column (to keep it a DataFrame)
Y = df['GDP per capita growth (annual %)']  # Replace column names as per your dataset

from sklearn.model_selection import train_test_split

# Split the data: 80% training, 20% testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

from sklearn.linear_model import LinearRegression

# Initialize and train the model
model = LinearRegression()
model.fit(X_train, Y_train)

# Display the coefficients
print(f"Intercept: {model.intercept_}")
print(f"Coefficient: {model.coef_}")

# Predict Y values for X_test
Y_pred = model.predict(X_test)

from sklearn.metrics import mean_squared_error, r2_score

# Calculate evaluation metrics
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")

import matplotlib.pyplot as plt

# Plot the scatter plot
plt.scatter(X, Y, color='#1fb462', alpha=0.6, label="Data points")

# Plot the regression line
plt.plot(X_test, Y_pred, color='#243bad', label=" 0.01310")

# Add labels and title
plt.xlabel("Ease of Doing Business (EDB)")
plt.ylabel("GDP Per Capita Growth (Annual %)")
plt.title("Regression between EDB and GDP per Capita Growth (2020)")
plt.legend()
plt.show()
```
